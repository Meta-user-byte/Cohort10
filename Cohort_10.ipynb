A standard way to represent the noise in a quantum circuit is through Pauli operators (x, y, z). Build a function with input p_single_error and p_double_error,  and QuantumCircuit of 2 qubits where:


*    p_single_error → Probability of having a random Pauli operator acting on the qubit after a one-qubit gate
*   Lisp_double-error → Probability of having a random Pauli operator acting on the qubit after a two-qubit gatet item


*   QuantumCircuit → Quantum circuit where the noise will be added
\
\
The output should be the Quantum Circuit with Noise

 **This code adds random Pauli noise to each qubit after a gate, based on specified probabilities for one-qubit (p_single) and two-qubit (p_double) gates, creating a new circuit (noisy_circuit) with simulated noise effects**

```python
from qiskit import QuantumCircuit
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister
import numpy as np

def apply_random_pauli(circuit, qubit):
    """Applies a random Pauli gate (X, Y, or Z) to the specified qubit."""
    random_pauli = np.random.choice(['x', 'y', 'z'])
    if random_pauli == 'x':
        circuit.x(qubit)
    elif random_pauli == 'y':
        circuit.y(qubit)
    elif random_pauli == 'z':
        circuit.z(qubit)

def add_noise(circuit, p_single_error, p_double_error):
    """Adds Pauli noise to a quantum circuit based on probabilities after each gate.

    Parameters:
    - circuit (QuantumCircuit): The quantum circuit where noise will be added.
    - p_single_error (float): Probability of Pauli noise after single-qubit gate.
    - p_double_error (float): Probability of Pauli noise after two-qubit gate.

    Returns:
    - noisy_circuit (QuantumCircuit): A new circuit with added noise.
    """
    noisy_circuit = QuantumCircuit(circuit.num_qubits)

    for instruction, qubits, _ in circuit:
        # Add the original instruction to the noisy circuit
        noisy_circuit.append(instruction, qubits)

        # Identify if it's a single-qubit or two-qubit gate
        if len(qubits) == 1 and np.random.rand() < p_single:
            # Apply random Pauli noise after a single-qubit gate
            apply_random_pauli(noisy_circuit, qubits[0])
        elif len(qubits) == 2 and np.random.rand() < p_double:
            # Apply random Pauli noise to each qubit after a two-qubit gate
            apply_random_pauli(noisy_circuit, qubits[0])
            apply_random_pauli(noisy_circuit, qubits[1])

    return noisy_circuit
```

```python
# Original Circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

# Adding Noise
p_single_error = 0.5  # 50% chance after single-qubit gate
p_double_error = 0.5  # 50% chance after two-qubit gate
noisy_qc = add_noise(qc, p_single_error, p_double_error)

# Display the noisy circuit
print(noisy_qc)

```


2.   **Gate Basis**


*   Quantum computers can implement only a set of gates that, with transformations, can represent any other possible gate. This set of gates is called the Gate Basis of the QPU. Build a function that transforms a general Quantum Circuit to the following gate basis: {CX,ID,RZ,SX,X}





In quantum computing, gate decomposition refers to the process of breaking down complex quantum gates into a sequence of simpler gates that are easier to implement on a quantum processor (QPU). This decomposition is crucial for ensuring that quantum circuits can be executed efficiently and correctly on quantum hardware, which may only support a limited set of gates.

Quantum computers are typically designed with a specific set of gates, called the gate basis, which forms the "native" set of gates that can be implemented directly. Gates outside this set must be decomposed into combinations of these native gates so that the quantum circuit can be executed on a quantum computer.





*   The gates that are already part of the basis {CX, ID, RZ, SX, X} should remain unchanged in the decomposition.

*   Hadamard (h): Decomposes as RZ(π) · SX · RZ(π).

*   u Gate: Decomposes as RZ(φ) · SX · RZ(θ) · SX · RZ(λ)

*  RX and RY Gates: Decompose into a combination of RZ and SX gates.

```python
from qiskit import QuantumCircuit
import numpy as np

def decompose_to_basis(circuit):
    """
    Decomposes a quantum circuit to the gate basis {CX, ID, RZ, SX, X}.
    """
    # Create a new quantum circuit to store the decomposed gates
    decomposed_circuit = QuantumCircuit(circuit.num_qubits)

    # Iterate over each gate in the original circuit
    for instruction, qubits, _ in circuit:
        name = instruction.name
        params = instruction.params

        if name == 'h':
            # Decompose Hadamard gate (H) to {RZ, SX}
            decomposed_circuit.rz(np.pi, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(np.pi, qubits[0])

        elif name == 'u':
            # Decompose U gate to {RZ, SX} (u(θ, φ, λ) -> RZ(φ) SX RZ(θ) SX RZ(λ))
            theta, phi, lam = params
            decomposed_circuit.rz(phi, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(theta, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(lam, qubits[0])

        elif name == 'rx':
            # Decompose RX gate to {RZ, SX}
            theta = params[0]
            decomposed_circuit.rz(np.pi, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(theta, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(np.pi, qubits[0])

        elif name == 'ry':
            # Decompose RY gate to {RZ, SX}
            theta = params[0]
            decomposed_circuit.rz(np.pi / 2, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(-np.pi / 2, qubits[0])
            decomposed_circuit.sx(qubits[0])
            decomposed_circuit.rz(theta, qubits[0])

        elif name == 'cx':
            # CX gate is already part of the target basis
            decomposed_circuit.cx(qubits[0], qubits[1])

        elif name == 'id':
            # ID gate is already part of the target basis
            decomposed_circuit.id(qubits[0])

        elif name == 'rz':
            # RZ gate is already part of the target basis
            decomposed_circuit.rz(params[0], qubits[0])

        elif name == 'sx':
            # SX gate is already part of the target basis
            decomposed_circuit.sx(qubits[0])

        elif name == 'x':
            # X gate is already part of the target basis
            decomposed_circuit.x(qubits[0])

        else:
            raise ValueError(f"Unsupported gate: {name}")

    return decomposed_circuit
```

```python
# Create a simple quantum circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.rx(np.pi/2, 0)
qc.cx(0, 1)  # A simple gate, but it could also be on multiple qubits

# Decompose the circuit to the target gate basis
decomposed_qc = decompose_to_basis(qc)

# Display the decomposed circuit
print(decomposed_qc)

```

*   Build a function (quantum_sum) to add two numbers using the Draper adder algorithm. You will need the Quantum Fourier Transform (QFT). Many libraries offer a function to use it. For this task, you will need to build QFT from scratch.






The Draper adder algorithm is a quantum algorithm for adding two binary numbers using the Quantum Fourier Transform (QFT)


  It involves three primary steps:
1. Quantum Fourier Transform (QFT): Transforms the input qubits into a superposition of states, allowing for parallel computations
2.   Controlled-Phase Gate Operations: These gates implement the addition logic by applying phase shifts to specific qubits based on the input values
3.   Inverse Quantum Fourier Transform (IQFT): Transforms the superposition back into a classical state, where the result of the addition is measured.



```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.circuit.library import QFT

def quantum_sum(a, b):
    """
    Adds two numbers using the Draper adder algorithm.

    Args:
        a: The first number.
        b: The second number.

    Returns:
        A QuantumCircuit representing the addition operation.
    """

    # Determine the number of qubits required
    num_qubits = max(len(bin(a)[2:]), len(bin(b)[2:]))

    # Create a quantum circuit
    qc = QuantumCircuit(num_qubits * 2, num_qubits)

    # Encode the numbers in qubits
    for i in range(num_qubits):
        if (a >> i) & 1:
            qc.x(i)
        if (b >> i) & 1:
            qc.x(i + num_qubits)

    # Apply QFT to the first register
    qc.append(QFT(num_qubits, inverse=False), range(num_qubits))

    # Controlled-Phase Gate Operations (Draper Adder logic)
    for control_qubit in range(num_qubits):
        for target_qubit in range(control_qubit + 1, num_qubits):
            angle = 2 * np.pi / 2**(target_qubit - control_qubit)
            qc.cp(angle, control_qubit, target_qubit)

    # Apply IQFT to the first register
    qc.append(QFT(num_qubits, inverse=True), range(num_qubits))

    # Measure the first register
    qc.measure(range(num_qubits), range(num_qubits))

    return qc

# Example usage:
a = 3
b = 5
qc = quantum_sum(a, b)
qc.draw() # Visualize the quantum circuit
```
